How does P265 behaviour vary with computer, and can we observe correlations among different computers? Which computers are most active in the change event?

We start by constructing the RDD of point processes, one for each computer, of P265 'Start' events. We throw away all other fields:

```python
p265_comps = proc_p265\
            .filter(lambda x: x[4]=='Start')\
            .map(lambda x: (x[3], x[0]))\
            .groupByKey()
```

Ideally, we should model these point processes. However, for simplicity we'll bin them into time series of hourly counts and examine the time series for each computer.

Conceptually, we store this as a matrix of counts with rows indexed by computer and columns indexed by hour up to hour 400.

Formally, we construct two RDD representations of this matrix, one sparse (since most counts are zero) and the other dense ('full'):

```python
# full frequency table of P265 starts by computer and time-bin up to hour 400:

from collections import Counter

binsize = 3600

def counter(dat):
    return Counter([t/binsize for t in dat])

def reduce(input, r=400 * 3600/binsize):
    c = input[0]
    d = input[1]
    for x in range(r):
        if x not in d.keys():
            d[x] = 0
    return [([c,k], v) for (k,v) in d.items() if k < r]
    
p265_comps_ts_sparse = p265_comps.map(lambda x : (x[0], counter(list(x[1]))))
p265_comps_ts_full = p265_comps_ts_sparse.flatMap(reduce)

p265_comps_ts_sparse.cache()
p265_comps_ts_full.cache()
```

Some typical time series (in sparse form, zero at all other hours):

```python
p265_comps_ts_sparse.take(3)
```

We see that the normal behaviour for P265 is at most a few starts per hour. The plots below explore the distribution of 'nr P265 starts' over the set of computers, in each of the hours from 338 to 344:  

```python
%matplotlib inline
import matplotlib.pyplot as plt

busyness = 300
start_hour = 338
end_hour = 345

for hour in range(start_hour, end_hour):
    hour_set = p265_comps_ts_sparse.filter(lambda x: x[1][hour] > 0).collect()
    vol_dist = [cnt[hour] for (cmp, cnt) in hour_set]
    nr_comp = len([cnt[hour] for (cmp, cnt) in hour_set if cnt[hour] > busyness])
    plt.hist(vol_dist, bins=20)
    plt.title("Hour %d: nr computers with >%d 'starts' = %d" % (hour, busyness, nr_comp))
    plt.show()
```

We see that the big change event occurs at hour 341: suddenly 2964 computers are exhibiting > 300 start events.

This starts to answer the question of which computers are directly involved in the change event. We'll return to this set of computers later. 

But first we ask how anomalous 'heavy hitter' computer-hours are: in the 'normal' period leading up to the change event, how often is any computer observed to have many start events? We rephrase this as asking for the distribution of values in the count matrix for column (hour) up to some number less than 341 (e.g. 300):

```python
# distribution of frequency table entries up to hour 330:

from math import log 

upper = 300

early_freq = p265_comps_ts_full\
            .filter(lambda x: x[0][1] <= upper)\
            .map(lambda x: x[1])\
            .countByValue()

fig = plt.figure(figsize=(20,8))
size = [100*log(1+y, 10) for y in early_freq.values()]
plt.xlim(xmin=7e-02, xmax=1e04)
plt.ylim(ymin=5e-01, ymax=1e07)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Nr P265 starts")
plt.ylabel("Frequency")
plt.title("Distribution of 'nr P265 starts' per computer-hour")
plt.scatter([x+0.1 for x in early_freq.keys()], early_freq.values(), s=size, alpha=0.3)
plt.show()
```

We see that even in this early period there are plenty of occurences of 'heavy hitter' computer-hours (the mass on the right). 

However, this is not a simple power law (try fitting a line to the low values on the left). The shift to the right indicates a second mode in the distribution even in this initial period.

Was P265 behaving anomalously in a small number of computers throughout the initial 14 days? In which case hour 341 represents a phase transition at which the anomaly is suddenly communicated to 3000 new computers.

The next plot shows the number of distinct heavy hitter computers for each hour through the period:

```python
"""
Construct set of heavy hitter computers (for some busy-ness threshold) 
in each time bin, and inspect time series of the size of this set:
"""

busyness = 200 * binsize/3600

heavy_hitter_sets = p265_comps_ts_full\
                    .filter(lambda x: x[1] > busyness)\
                    .map(lambda x: (x[0][1], x[0][0]))\
                    .groupByKey()\
                    .map(lambda x: (x[0], set(x[1])))\
                    .collect()

x = [c[0] for c in heavy_hitter_sets]
y = [len(c[1]) for c in heavy_hitter_sets]

plt.figure(figsize=(20,8))
plt.xlabel("Hour")
plt.ylabel("Nr P265 heavy hitters")
plt.title("Time series of nr P265 heavy hitter computers")
plt.yscale("log")
plt.plot(x,y)
plt.show()
```

Finally, are these different computers or the same 're-offending' in each succesive hour? The plot below shows the cumulative set of distinct computers with time, leading up to the explosive transition at hour 341:

```python
lookup = dict(heavy_hitter_sets)
heavy_hitter_cum_set = dict()

T = 400 * 3600/binsize # time range

heavy_hitter_cum_set[0] = set([])
for t in range(1,T):
    if t not in lookup.keys():
        lookup[t] = set([])
    heavy_hitter_cum_set[t] = heavy_hitter_cum_set[t-1].union(lookup[t])

x = heavy_hitter_cum_set.keys()
heavy_hitter_cum_count = [len(heavy_hitter_cum_set[t]) for t in x]

plt.figure(figsize=(20,8))
plt.xlabel("Hour")
plt.ylabel("Nr heavy hitters")
plt.title("Cumulative number of P265 heavy hitter computers")
plt.stem(x, heavy_hitter_cum_count)
plt.show()
```

This allows us to identify the time sequence in which computers are added to the anomalous set. Let's do this and dump to .json format for later use:

```python
import json

computers = procs.map(lambda x: x[3]).distinct().collect()
computers = set(computers)
computer_p265_hh_hour = dict([(c, 2000) for c in computers])

for h in heavy_hitter_cum_set.keys():
    for c in heavy_hitter_cum_set[h]:
        if computer_p265_hh_hour[c] > h:
            computer_p265_hh_hour[c] = h
            
filename = lanl_path + "computer_p265_hh_dict.json"
fp = open(filename, 'w')
json.dump(computer_p265_hh_hour, fp)
fp.close()
```

<h2>Conclusions</h2>

We now learnt that P265 is a sparsely occuring process with a low background start-rate - mostly less than 10 per hour and usually zero. However, the 14-day period up to the change event is characterised by the gradual appearance of 'heavy hitters': computers-hours with significantly more P265 starts.

We defined a heavy hitter as at least 200 starts in an hour, loosely on the basis of the skewed power law distribution of values in the computer-hour count matrix.

The change event at hour 341 is characterised addition of 3000 new heavy hitter computers.

For later use, we label each computer in the network with the hour at which it becomes (if at all) a P265 heavy hitter.

<h2>ADAP: contingency tables</h2>

As in the previous notebook, this analysis consists of further conditioning of the process data set: having restricted to P265 we drop the user and domain fields (user is always root so contains no information; we drop domain for convenience), and one field 'Start/End' is binary and we restrict to the value 'Start'.

This reduces us to a point process (i.e. sequence of timestamps) for each value of the 'computer' field. We should ideally use the full information in these point processes - but for simplicity (and as usual) we bin up into time series with 1-hour resolution.

This reduces the analysis to two categorical variables (computer and time bin), which we study via the matrix of counts (or contingency table).
